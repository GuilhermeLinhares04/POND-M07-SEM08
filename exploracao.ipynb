```python
import pandas as pd
```

```python
# Convertendo dados para csv
df = pd.read_excel('dados/Bitcoin_07_09_2023-06_09_2024_historical_data_coinmarketcap.xlsx')
df.to_csv('dados/Bitcoin_1Y.csv', index=False)
df
```

```python
# Verificando o tipo de dados
df.info()
```

```python
# Normalizando valores inteiros para os preços
for col in df.select_dtypes(include=['int64']).columns:
    df[col] = df[col] // 10**10  # Dividindo por 10^10 e descartando os últimos 10 dígitos

df
```

```python
# Verificar número de dados da coluna 'open'
df['open'].value_counts()
```

```python
# Função para multiplicar por 10 os valores menores que 10 mil
for col in df.select_dtypes(include=['int64']).columns:
    df[col] = df[col].apply(lambda x: x * 10 if x < 10000 else x)


df.to_csv('dados/Bitcoin_1Y_Normalizado.csv', index=False)
df
```

```python
# Convertendo a coluna 'date' para datetime
df['timeOpen'] = pd.to_datetime(df['timeOpen'])
df['timeClose'] = pd.to_datetime(df['timeClose'])
df['timeHigh'] = pd.to_datetime(df['timeHigh'])
df['timeLow'] = pd.to_datetime(df['timeLow'])
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Ordem cronológica pela coluna 'timestamp'
df = df.sort_values(by='timestamp')

df
```

## Primeiro teste com modelo ARIMA

O modelo ARIMA é um modelo estatístico que utiliza a autocorrelação dos dados para prever valores futuros. Para este projeto, ele foi escolhido por ser um modelo simples, porém eficaz para séries temporais. Além disso, ele pode ser considerado versátil e flexível pela facilidade de ajustar os parâmetros do modelo. Por fim, suas previsões são baseadas em valores passados, o que é ideal para prever o possíveis valores de compras e vendas de ativos como é o caso deste projeto.

```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Prepara a série de preços 'close' para ARIMA
close_prices = df['close']

# Dividindo os dados em treinamento e teste (90% para treinamento)
train_size = int(len(close_prices) * 0.9)
train_data, test_data = close_prices[:train_size], close_prices[train_size:]

# Modelo ARIMA nos dados de treinamento
model = ARIMA(train_data, order=(5,1,2)) 
arima_result = model.fit()

# Realizar previsões
predictions = arima_result.forecast(steps=len(test_data))

# Plotando os valores reais vs previstos
plt.figure(figsize=(10,6))
plt.plot(train_data.index, train_data, label='Train')
plt.plot(test_data.index, test_data, label='Test', color='green')
plt.plot(test_data.index, predictions, label='Predicted', color='red')
plt.legend(loc='best')
plt.title('ARIMA Model: Actual vs Predicted Close Prices')
plt.show()
```

## Segundo teste com modelo ARIMA

```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
import joblib

# Prepara a série de preços 'close' para ARIMA
close_prices = df['close']

# Dividindo os dados em treinamento e teste (70% para treinamento)
train_size = int(len(close_prices) * 0.7)
train_data, test_data = close_prices[:train_size], close_prices[train_size:]

# Modelo ARIMA nos dados de treinamento
model = ARIMA(train_data, order=(10,0,0))
arima_result = model.fit()

# Realizar previsões
predictions = arima_result.forecast(steps=len(test_data))

# Plotando os valores reais vs previstos
plt.figure(figsize=(10,6))
plt.plot(train_data.index, train_data, label='Train')
plt.plot(test_data.index, test_data, label='Test', color='green')
plt.plot(test_data.index, predictions, label='Predicted', color='red')
plt.legend(loc='best')
plt.title('ARIMA Model: Actual vs Predicted Close Prices')
plt.show()

# Salvando o modelo ARIMA
joblib.dump(arima_result, 'models/arima_model.pkl')
```
